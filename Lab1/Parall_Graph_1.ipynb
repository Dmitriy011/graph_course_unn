# –ú–æ–¥–µ–ª—å –≠—Ä–¥—ë—à–∞ ‚Äî –†–µ–Ω—å–∏ G(n,p)

 **–ù–µ–æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ** –≥—Ä–∞—Ñ—ã G  
 **p - –ø–∞—Ä–∞–º–µ—Ç—Ä:** –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è —Ä–µ–±—Ä–∞ –º–µ–∂–¥—É –¥–≤—É–º—è –≤–µ—Ä—à–∏–Ω–∞–º–∏:  

### –ü–ª–æ—Ç–Ω–æ—Å—Ç—å: 

**–ï—Å–ª–∏ ùëù = 1:** –≥—Ä–∞—Ñ –±—É–¥–µ—Ç –ø–æ–ª–Ω—ã–º (–º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –ø–ª–æ—Ç–Ω—ã–º).  
**–ï—Å–ª–∏ ùëù = 0:** –≥—Ä–∞—Ñ –±—É–¥–µ—Ç –ø—É—Å—Ç—ã–º (–±–µ–∑ —Ä—ë–±–µ—Ä, –º–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å).  
**–ó–Ω–∞—á–µ–Ω–∏—è ùëù –º–µ–∂–¥—É 0 –∏ 1:** —á–µ–º –±–æ–ª—å—à–µ ùëù  - —Ç–µ–º –±–æ–ª—å—à–µ —Ä—ë–±–µ—Ä –±—É–¥–µ—Ç –¥–æ–±–∞–≤–ª–µ–Ω–æ –≤ –≥—Ä–∞—Ñ -> –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–æ–∑—Ä–∞—Å—Ç—ë—Ç.

## ‚úÖ –ü–æ–¥–∫–ª—é—á–µ–Ω–∏–µ –ø–∞–∫–µ—Ç–æ–≤:

1) **NetworkX** ‚Äî —ç—Ç–æ –±–∏–±–ª–∏–æ—Ç–µ–∫–∞, –ø—Ä–µ–¥–Ω–∞–∑–Ω–∞—á–µ–Ω–Ω–∞—è –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è, –∞–Ω–∞–ª–∏–∑–∞ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ —Å–ª–æ–∂–Ω—ã—Ö —Å–µ—Ç–µ–π –∏ –≥—Ä–∞—Ñ–æ–≤  
2)  **Matplotlib.pyplot** - –Ω–∞—É—á–Ω–∞—è –≥—Ä–∞—Ñ–∏–∫–∞

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## ‚úÖ –ü–æ–¥—Å—á–µ—Ç —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –≥—Ä–∞—Ñ–∞

```python
def calculate_Graph_characteristics(G):
    # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≥—Ä–∞—Ñ–∞
    density = nx.density(G)
    print(f"–ü–ª–æ—Ç–Ω–æ—Å—Ç—å: {density}")

    # –°–≤—è–∑–Ω–æ—Å—Ç—å
    is_connected = nx.is_connected(G)
    largest_component = max(nx.connected_components(G), key=len)
    print(f"–°–≤—è–∑–Ω–æ—Å—Ç—å: {is_connected}")
    print(f"–†–∞–∑–º–µ—Ä –Ω–∞–∏–±–æ–ª—å—à–µ–π –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã: {len(largest_component)}")

    # –î–∏–∞–º–µ—Ç—Ä
    diameter = nx.diameter(G)
    print(f"–î–∏–∞–º–µ—Ç—Ä: {diameter}")

    # –î–ª–∏–Ω–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫—Ä–∞—Ç—á–∞–π—à–µ–≥–æ –ø—É—Ç–∏
    average_shortest_path_length = nx.average_shortest_path_length(G)
    print(f"–°—Ä–µ–¥–Ω—è—è –¥–ª–∏–Ω–∞ –∫—Ä–∞—Ç—á–∞–π—à–µ–≥–æ –ø—É—Ç–∏: {average_shortest_path_length}")

    # –°—Ä–µ–¥–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏
    average_clustering = nx.average_clustering(G)
    print(f"–°—Ä–µ–¥–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏: {average_clustering}")
    
    # –ü–æ–ª—É—á–∞–µ–º —Å—Ç–µ–ø–µ–Ω–∏ –≤—Å–µ—Ö —É–∑–ª–æ–≤
    degree_sequence = [degree for node, degree in G.degree()]
    plt.hist(degree_sequence, bins=50)
    plt.title("Degree Distribution")
    plt.xlabel("Degree")
    plt.ylabel("Count")
    plt.show()
```

## ‚úÖ –ì—Ä–∞—Ñ—ã —Ä–∞–∑–º–µ—Ä–∞ n = 10 —Å —Ä–∞–∑–Ω–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç—å—é –ø—Ä–∏ p = 1; 0.9; 0.4 

```python
# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –≥—Ä–∞—Ñ–æ–≤
n = 10  
p_values = [1, 0.9, 0.4]

# –°–æ–∑–¥–∞–Ω–∏–µ —Ñ–∏–≥—É—Ä—ã –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–æ–≤
fig, axes = plt.subplots(1, len(p_values), figsize=(15, 5))  # –æ–¥–∏–Ω —Ä—è–¥, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–æ–ª–±—Ü–æ–≤ = –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≥—Ä–∞—Ñ–æ–≤
G1, G2, G3 = [], [], []

for i, p in enumerate(p_values):
    # –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –≥—Ä–∞—Ñ–∞ –≠—Ä–¥—ë—à–∞-–†–µ–Ω—å–∏
    G = nx.erdos_renyi_graph(n, p)
    if(i == 0): G1 = G
    if(i == 1): G2 = G
    if(i == 2): G3 = G
    
    # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≥—Ä–∞—Ñ–∞
    density = nx.density(G)
    
    # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–π –æ—Å–∏
    ax = axes[i]
    nx.draw(G, with_labels = True, node_color = "lightblue", node_size = 500, font_size = 10, ax = ax)
    
    # –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ —Å –ø–æ–∫–∞–∑–æ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ p –∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏
    ax.set_title(f"p = {p}, –ü–ª–æ—Ç–Ω–æ—Å—Ç—å = {density:.2f}")

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –≤—Å–µ –≥—Ä–∞—Ñ—ã
plt.show()
```

```python
calculate_Graph_characteristics(G1)
```

```python
calculate_Graph_characteristics(G2)
```

```python
calculate_Graph_characteristics(G3)
```

## ‚úÖ –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –≥—Ä–∞—Ñ—ã 4–µ—Ö —Ä–∞–∑–Ω—ã—Ö —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–µ–π –∏ 3–µ—Ö –≤–∏–¥–æ–≤ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ 

### üí¢ $G_{1}$:  $n = 100000$ –∏ $p = 0.000124$

```python
n = 100000  # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω
p = 0.000124  # –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è —Ä–µ–±—Ä–∞

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –≥—Ä–∞—Ñ–∞ –≠—Ä–¥—ë—à–∞-–†–µ–Ω—å–∏
G_1 = nx.erdos_renyi_graph(n, p)
```

```python
num_edges = G_1.number_of_edges() 
print(num_edges) 
```

### 1Ô∏è‚É£ –ü–ª–æ—Ç–Ω–æ—Å—Ç—å

–ù–µ–æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≥—Ä–∞—Ñ   
**ùëõ:** –≤–µ—Ä—à–∏–Ω   
**ùëö:** —Ä—ë–±–µ—Ä   

$$ D = \frac{2m}{n(n-1)} $$

```python
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≥—Ä–∞—Ñ–∞
density = nx.density(G_1)
print(f"–ü–ª–æ—Ç–Ω–æ—Å—Ç—å: {density}")
```

### 2Ô∏è‚É£ –°–≤—è–∑–Ω–æ—Å—Ç—å

**–°–≤—è–∑–Ω—ã–π –≥—Ä–∞—Ñ** - —ç—Ç–æ –≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ø—É—Ç—å –º–µ–∂–¥—É –ª—é–±–æ–π –ø–∞—Ä–æ–π –≤–µ—Ä—à–∏–Ω.  

```python
# –°–≤—è–∑–Ω–æ—Å—Ç—å
is_connected = nx.is_connected(G_1)
largest_component = max(nx.connected_components(G_1), key=len)
print(f"–°–≤—è–∑–Ω–æ—Å—Ç—å: {is_connected}")
print(f"–†–∞–∑–º–µ—Ä –Ω–∞–∏–±–æ–ª—å—à–µ–π –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã: {len(largest_component)}")
```

### 3Ô∏è‚É£ –î–∏–∞–º–µ—Ç—Ä

**–î–∏–∞–º–µ—Ç—Ä –≥—Ä–∞—Ñ–∞** ‚Äî —ç—Ç–æ –Ω–∞–∏–±–æ–ª—å—à–∞—è –¥–ª–∏–Ω–∞ –∫—Ä–∞—Ç—á–∞–π—à–µ–≥–æ –ø—É—Ç–∏ –º–µ–∂–¥—É –¥–≤—É–º—è –≤–µ—Ä—à–∏–Ω–∞–º–∏ –≤ –≥—Ä–∞—Ñ–µ.  
'#' –ï—Å–ª–∏ –≥—Ä–∞—Ñ –Ω–µ—Å–≤—è–∑–Ω—ã–π, –¥–∏–∞–º–µ—Ç—Ä –æ–±—ã—á–Ω–æ —Å—á–∏—Ç–∞–µ—Ç—Å—è –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–º, –ø–æ—Ç–æ–º—É —á—Ç–æ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ø—É—Ç–∏ –º–µ–∂–¥—É –Ω–µ–∫–æ—Ç–æ—Ä—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏.

```python
# –î–∏–∞–º–µ—Ç—Ä

if(is_connected):
    diameter = nx.diameter(G_1)
else:
    diameter = -1

print(f"–î–∏–∞–º–µ—Ç—Ä: {diameter}")
```

### 4Ô∏è‚É£ –î–ª–∏–Ω–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫—Ä–∞—Ç—á–∞–π—à–µ–≥–æ –ø—É—Ç–∏

**–î–ª–∏–Ω–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫—Ä–∞—Ç—á–∞–π—à–µ–≥–æ –ø—É—Ç–∏** ‚Äî —ç—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤—Å–µ—Ö –∫—Ä–∞—Ç—á–∞–π—à–∏—Ö –ø—É—Ç–µ–π –º–µ–∂–¥—É –≤—Å–µ–º–∏ –ø–∞—Ä–∞–º–∏ –≤–µ—Ä—à–∏–Ω –≤ –≥—Ä–∞—Ñ–µ  
$$\frac{1}{n(n-1)} \sum_{u, v \in V, u \neq v} d(u, v)$$  
**d(u,v):** –¥–ª–∏–Ω–∞ –∫—Ä–∞—Ç—á–∞–π—à–µ–≥–æ –ø—É—Ç–∏ –º–µ–∂–¥—É u –∏ v

```python
# –î–ª–∏–Ω–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫—Ä–∞—Ç—á–∞–π—à–µ–≥–æ –ø—É—Ç–∏
average_shortest_path_length = nx.average_shortest_path_length(G_1)
print(f"–°—Ä–µ–¥–Ω—è—è –¥–ª–∏–Ω–∞ –∫—Ä–∞—Ç—á–∞–π—à–µ–≥–æ –ø—É—Ç–∏: {average_shortest_path_length}")
```

### 5Ô∏è‚É£ –°—Ä–µ–¥–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏

**–°—Ä–µ–¥–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ (Average Clustering Coefficient)** - —ç—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –≤—Å–µ—Ö –≤–µ—Ä—à–∏–Ω –≥—Ä–∞—Ñ–∞:  
**–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏** –∏–∑–º–µ—Ä—è–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ, —á—Ç–æ –¥–≤–∞ —Å–æ—Å–µ–¥–∞ –æ–¥–Ω–æ–π –≤–µ—Ä—à–∏–Ω—ã —Ç–∞–∫–∂–µ —Å–æ–µ–¥–∏–Ω–µ–Ω—ã –º–µ–∂–¥—É —Å–æ–±–æ–π (–æ–±—Ä–∞–∑—É—é—Ç —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫).  
$$G_{avg} = \frac{1}{n}\sum_{v \in V} G_{v}$$  
**'#'** –Ω–∞—Å–∫–æ–ª—å–∫–æ –≤–µ—Ä—à–∏–Ω—ã —Å–∫–ª–æ–Ω–Ω—ã –æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –≥—Ä—É–ø–ø—ã (–∏–ª–∏ ¬´–∫–ª–∞—Å—Ç–µ—Ä—ã¬ª).

```python
# –°—Ä–µ–¥–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏
average_clustering = nx.average_clustering(G_1)
print(f"–°—Ä–µ–¥–Ω–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏: {average_clustering}")
```

### 6Ô∏è‚É£ –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–µ–ø–µ–Ω–µ–π

```python
# –ü–æ–ª—É—á–∞–µ–º —Å—Ç–µ–ø–µ–Ω–∏ –≤—Å–µ—Ö —É–∑–ª–æ–≤
degree_sequence = [degree for node, degree in G_1.degree()]
plt.hist(degree_sequence, bins = 50)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.show()
```
